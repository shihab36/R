```r
#Making a #D pieChart
install.packages("plotrix")
library(plotrix)
```

```r
person = c(10,10,5,5,10)                                                                                                                                                                
sports = c("Badmintion","Cricket","Volyball","Football","Hokey")  
```

```r
pie(person, sports, main = "Favorite Sports Percentage", col = rainbow(length(person)))                                                             
legend("right", sports, cex = .60, fill = rainbow(length(person)))                                       
```

```r
pie3D (person, labels=sports, explode =.05, main = "Favourite Sports Percentage")
```
